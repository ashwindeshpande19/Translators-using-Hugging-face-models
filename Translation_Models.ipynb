Merging two datasets

```python
# Load English DataFrame
df_english = pd.read_csv('flores200_data_English.csv')

# Load Hindi DataFrame
df_hindi = pd.read_csv('flores200_data_hin_Deva.csv')

# Merge DataFrames based on common columns
merged_df = pd.merge(df_english, df_hindi, on=['id', 'domain', 'topic'])

# Rename 'sentence' columns
merged_df.rename(columns={'sentence_x': 'English_Sentence', 'sentence_y': 'Hindi_Sentence'}, inplace=True)

# Save the merged DataFrame to a new CSV file
merged_csv_filename = 'merged_flores200_data.csv'
merged_df.to_csv(merged_csv_filename, index=False)
print(f"Data has been saved to {merged_csv_filename}")

import pandas as pd

# Load the merged CSV file into a new DataFrame
merged_df = pd.read_csv('merged_flores200_data.csv')

# Print the first 5 rows of the merged DataFrame for verification
print(merged_df.head())
```

```python
# Print the columns of the merged DataFrame
print(merged_df.columns)
```

```python
# Drop unwanted columns
columns_to_drop = ['URL_x', 'has_image_x', 'has_hyperlink_x', 'URL_y', 'has_image_y', 'has_hyperlink_y']
merged_df = merged_df.drop(columns=columns_to_drop)

# Print the updated columns of the merged DataFrame
print(merged_df.columns)
```

Downloading Helsinki-NLP/opus-mt-en-hi model

```python
from transformers import pipeline, AutoTokenizer, AutoModelForSeq2SeqLM
import pickle
translator_opus = pipeline("translation_en_to_hi", model="Helsinki-NLP/opus-mt-en-hi", device=-1)
# Input an English sentence
english_sentence = input("Enter an English sentence: ")
# Translate the input sentence into Hindi using different models
hindi_translation_opus = translator_opus(english_sentence, max_length=50)[0]['translation_text']
# Print the translations
print("Translation (opus):", hindi_translation_opus)
```

Downloading facebook/m2m100_418M model

```python
from transformers import pipeline, AutoTokenizer, AutoModelForSeq2SeqLM
import pickle
translator_m2m100 = pipeline("translation_en_to_hi", model="facebook/m2m100_418M", device=-1)
# Input an English sentence
english_sentence = input("Enter an English sentence: ")
hindi_translation_m2m100 = translator_m2m100(english_sentence, max_length=50)[0]['translation_text']
print("Translation (m2m100):", hindi_translation_m2m100)
```

Downloading facebook/mbart-large-50-many-to-many-m model

```python
from transformers import pipeline, AutoTokenizer, AutoModelForSeq2SeqLM
import pickle
translator_mbart = pipeline("translation", model="facebook/mbart-large-50-many-to-many-mmt", device=-1, src_lang="en_XX", tgt_lang="hi_IN")
# Input an English sentence
english_sentence = input("Enter an English sentence: ")
hindi_translation_mbart = translator_mbart(english_sentence, max_length=50)[0]['translation_text']
print("Translation (mbart):", hindi_translation_mbart)
```

BLEU Score

```python
from sacrebleu import corpus_bleu

# Assuming you have reference translations in Hindi for your dataset
reference_translations = merged_df_subset['Hindi_Sentence'].tolist()

# Calculate BLEU score for each model
translations_opus = merged_df_subset['Hindi_Translation_opus'].tolist()
translations_m2m100 = merged_df_subset['Hindi_Translation_m2m100'].tolist()
translations_mbart = merged_df_subset['Hindi_Translation_mbart'].tolist()

# Convert each element to string
translations_opus = [str(x) for x in translations_opus]
translations_m2m100 = [str(x) for x in translations_m2m100]
translations_mbart = [str(x) for x in translations_mbart]

# Calculate BLEU score for each translation
bleu_opus = corpus_bleu(translations_opus, [reference_translations])
bleu_m2m100 = corpus_bleu(translations_m2m100, [reference_translations])
bleu_mbart = corpus_bleu(translations_mbart, [reference_translations])

# Print the BLEU scores
print(f'BLEU Score for Opus-MT: {bleu_opus.score}')
print(f'BLEU Score for m2m100: {bleu_m2m100.score}')
print(f'BLEU Score for mbart: {bleu_mbart.score}')
```
